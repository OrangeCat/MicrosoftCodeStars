```python
factorial = lambda x: factorial(x - 1) * x if x > 1 else 1
factorial(104)*10//3600
```

```python
def fact(n):
    f=1
    for i in range(1, n+1):
        f=f*i
    return f

print (fact(104)*10//60//60)
```
